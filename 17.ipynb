```
from itertools import combinations
import pandas as pd
```

```
data = pd.DataFrame({'score': {1: 78, 2: 64, 3: 68, 4: 62, 5: 85},
                     'workers': {1: 7, 2: 4, 3: 5, 4: 5, 5: 8}})
max_workers = 23

data.head()
```

```
prior = None
max_score = 0
for i in range(1, len(data)+1):
    for indexes in combinations(data.index, i):
        var = data[data.index.isin(indexes)]
        if sum(var['workers']) > max_workers:
            continue
        
        if max_score < sum(var['score']):
            max_score = sum(var['score'])
            prior = var

print(f'Max score: {max_score}')
print(f'Заявки к одобрению: {[_ for _ in prior.index]}')
print(prior)
```
